```python
class Dog_Dataset(Dataset):

    def __init__(self,ds,transform=None):
        self.ds = ds
        self.transform = transform

    def __len__(self):
        return len(self.ds)

    def __getitem__(self,idx):
        image,label = self.ds[idx]
        if self.transform:
            image = self.transform(image)
            return image,label
```

```python
#mean_std = ([0.485,0.456,0.406],[0.229,0.224,0.225])

train_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomCrop(224,padding=4,padding_mode="reflect"),
    transforms.RandomHorizontalFlip(p=0.3),
    transforms.RandomRotation(degrees=30),
    transforms.ToTensor()
])


val_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

test_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
])
```

```python
train_dataset = Dog_Dataset(train_ds,train_transform)
valid_dataset = Dog_Dataset(val_ds,val_transform)
test_dataset= Dog_Dataset(test_ds,test_transform)
```

```python
def accuracy(outputs,labels):
    _,preds = torch.max(outputs,dim=1)
    return torch.tensor(torch.sum(preds==labels).item()/len(preds))
```

```python
class ImageClassification(nn.Module):

    def training_step(self,batch):
        image,targets = batch
        out = self(image)
        loss = F.nll_loss(out,targets)
        return loss

    def validation_step(self,batch):
        image,targets = batch
        out = self(image)
        loss = F.nll_loss(out,targets)
        acc = accuracy(out,targets)
        return {"val_acc":acc.detach(),"val_loss":loss.detach()}

    def validation_epoch_end(self,outputs):
        batch_losses = [x["val_loss"] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()
        batch_accs = [x["val_acc"] for x in outputs]
        epoch_acc = torch.stack(batch_accs).mean()
        return {"val_loss":epoch_loss.item(),"val_acc":epoch_acc.item()}

    def epoch_end(self,epoch,result):
        print("Epoch = [{}], train_loss: {:.4f}, val_loss={:.4f}, val_acc={:.4f}".format(epoch,result["train_loss"],result["val_loss"],result["val_acc"]))
```

```python
class Dog_Breed_Classification_Model(ImageClassification):
    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(

            nn.Conv2d(3,32,3,stride=1,padding=1),
            nn.ReLU(),
            nn.Conv2d(32,32,3,stride=1,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(32,64,3,stride=1,padding=1),
            nn.ReLU(),
            nn.Conv2d(64,128,3,stride=1,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(128,224,3,stride=1,padding=1),
            nn.ReLU(),
            nn.Conv2d(224,224,3,stride=1,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(224,224,3,stride=1,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(224,224,3,stride=1,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Flatten(),
            nn.Linear(7*7*224,512),
            nn.ReLU(),
            nn.Linear(512,120),
            nn.LogSoftmax(dim=1),

        )


    def forward(self,xb):
        return self.network(xb)
```

```python
model = Dog_Breed_Classification_Model()
model
```

```python
class DogBreedPretrainedResnet34(ImageClassification):
  def __init__(self):
    super().__init__()

    self.network = models.resnet34(pretrained=True)

    num_ftrs = self.network.fc.in_features
    self.network.fc = nn.Sequential(
        nn.Linear(num_ftrs,120),
        nn.LogSoftmax(dim=1)
    )
  def forward(self,xb):
    return self.network(xb)
```

```python
def get_default_device():
    if torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")

def to_device(data,device):
    if isinstance(data,(list,tuple)):
        return [to_device(d,device) for d in data]
    else:
        return data.to(device,non_blocking=True)
```

```python
class DeviceDataloader:
    def __init__(self,dl,device):
        self.dl = dl
        self.device = device

    def __len__(self):
        return len(self.dl)

    def __iter__(self):
        for batch in self.dl:
            yield to_device(batch,self.device)
```

```python
device = get_default_device()
print(device)
```

```python
from tqdm.notebook import tqdm
```

```python
def get_learning_rate(optimizer):
    for param_group in optimizer.param_groups:
        return param_group["lr"]

def fit_one_cycle(epochs,max_lr,model,train_loader,valid_loader,weight_decay=0,grad_clip=None,opt_func=torch.optim.Adam):
    torch.cuda.empty_cache()
    history = []
    optimizer = opt_func(model1.parameters(),max_lr,weight_decay = weight_decay)

    sched = torch.optim.lr_scheduler.OneCycleLR(optimizer,max_lr,epochs=epochs,steps_per_epoch = len(train_loader))


    for epoch in range(epochs):
        model1.train()
        train_losses=[]
        lrs = []

        for batch in tqdm(train_loader):
            loss = model1.training_step(batch)
            train_losses.append(loss)

            loss.backward()


            optimizer.step()
            optimizer.zero_grad()

            lrs.append(get_learning_rate(optimizer))

            sched.step()

        result = evaluate(model1,valid_loader)
        result["train_loss"] = torch.stack(train_losses).mean().item()
        result["lrs"] = lrs
        model1.epoch_end(epoch,result)
        history.append(result)

    return history


@torch.no_grad()
def evaluate(model1,valid_loader):
    model1.eval()
    output = [model1.validation_step(batch) for batch in valid_dataloader]
    return model1.validation_epoch_end(output)
```
